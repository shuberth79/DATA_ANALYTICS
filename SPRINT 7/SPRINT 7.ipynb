```python
"""
EJERCICIO 1__________________________________________________________________
Calculadora del índice de masa corporal
Escribe una función que calcule el IMC ingresado por el usuario/a, es decir, 
quien lo ejecute deberá ingresar estos datos. 
"""
```

```python
# En primera instancia, casi en todos los ejercicios hemos incluido un detalle personalizado para la bienvenida
# y trato con el usuario, posterior a ello definimos las funciones:
# (A) función_inicial: solicitamos (input) las dimensiones de estatura y peso como tipo de dato (FLOAT),
    # luego lo guardamos con RETURN
# (B) función calcula_icm: las dimensiones de estatura y peso los condicionamos (IF) si son menosres de '0' salte un
    # anuncio de error y se retorne al inicio de la operación, desestimando de esta forma a los numeros negativos.
# (C) función clasifica_icm: esta parte es muy impotr6ante, ya que despues de filtrar, categorizamos las entradas 
    # los indices de 'icm' con (IF, ELIF, ELSE)
# (D) función pregunta_usuario: este es un plus que usualmente se lo incopora, para lograr operatividad ciclica
# (E) función main: definimos el resumen y orden de las ejecuciones dentro de un bucle hasta que el usuario
    # desista continuar (BREAK)
    
print('-'*55)
print('>>>> CALCULADORA DEL INDICE DE MASA CORPORAL (IMC) <<<<')
print('-'*55)
name = input("Estimado usuario cual es su nombre?: ")
Name = name.title()

def funcion_inicial():   #--------(A)
    altura = float(input("Por favor ingrese su estatura (metros): "))
    peso = float(input("Por favor ingrese su peso (kilogramos): "))
    return altura, peso

def calcula_imc(altura, peso):  #--------(B)
    if altura <= 0:
        raise ValueError("La altura debe ser un valor positivo.")
    imc = peso / (altura * altura)
    return imc

def clasifica_imc(imc):   #--------(C)
    if imc < 18.5:
        return "indice de Bajo peso"
    elif imc < 25:
        return "indice de Peso normal"
    elif imc < 30:
        return "indice de Sobrepeso"
    else:
        return "indice de Obesidad"

def pregunta_usuario():     #--------(D)
    respuesta = input("¿Desea calcular otro Índice Corporal? (Si/No): ").lower()
    return respuesta in ["si", "sí"]

def main():     #--------(E)
    while True:
        altura, peso = funcion_inicial()
        imc = calcula_imc(altura, peso)
        clasificacion = clasifica_imc(imc)
        print('-------------------------------------------------------')
        print(f"Estimado/a {Name} su índice de masa corporal (IMC) es: {imc:.2f}")
        print(f"Clasificación: {clasificacion}")

        if not pregunta_usuario():
            break

if __name__ == "__main__":
    main()
```

```python
"""
EJERCICIO 2__________________________________________________________________
Convertidor de temperatures.
Existen diversas unidades de temperatura utilizadas en distintos contextos y regiones.
Las más comunes son Celsius (°C), Fahrenheit (°F) y Kelvin (K). 
También existen otras unidades como Rankine (°Ra) y Réaumur (°Re). 
Selecciona al menos 2 y construye el conversor.
"""
```

```python
# (A) función pregunta_usuario: hemos incluido un detalle personalizado para preguntar al usuario si desea
    # reiterar la operacion
# (B) función conversor_temperatura: es la patrte central del script, en 'unidad_inicial' recibimos las entradas
    # por (INPUT) ingresando las iniciales de: Celsius(C), Fahrenheit(F) o Kelvin(K), lo filtramos con (IF),
    # si el ingreso no es válido saltará un error con (RAISE), si es válido pasará a otro condicional (IF, ELIF, ELSE)
    # a las entradas de 'unidad_inicial' y 'unidad_final' para definir que unidades se van a convertir y que unidades
    # desea obtener, de esta forma se induce a la conversión numérica, conforme la fórmula aritmética equivalente.
    # ya sea de Celsius a Kelvin o Fahrenheit, de Kelvin a Celsius o Fahrenheit, de Fahrenheit a Kelvin o Celsius
# (C) EXCEPT: ejecuta función pregunta_usuario, aqui el usuario decide si desea continuar usando el convertidor o no.

print('-----------------------------------------------')
print('    >>> CONVERTIDOR - DE - TEMPERATURA <<<')
print('-----------------------------------------------')
print('--- Celsius(C) - Fahrenheit(F) - Kelvin(K)---')
nombre = input("Indique cual es su nombre?: ")
Nombre = nombre.title()

def pregunta_usuario():     #--------(A)
    while True:
        respuesta = input("¿Desea realizar otra conversión? (Si/No): ").lower()
        if respuesta in ["si", "sí"]:
            print("Ok, continuemos!!!")
            return True
        elif respuesta in ["no", "no"]:
            print("Perfecto, Cuidate, adiós")
            return False
        else:
            print("Ingrese 'Si' o 'No' para continuar o salir.")

def conversor_temperatura():     #--------(B)
    while True:
        try:                
            unidad_inicial = input("Indique la unidad de temperatura que desea convertir: Celsius(C), Fahrenheit(F) o Kelvin(K): ").upper()
            if unidad_inicial not in ['C', 'K', 'F']:
                raise ValueError("Información ingresada no válida. Ingrese C, F o K.")

            unidad_final = input("Indique la unidad de temperatura que desea obtener: ").upper()
            if unidad_final not in ['C', 'K', 'F']:
                raise ValueError("Información ingresada no válida. Ingrese C, F o K.")

            temperatura = float(input("Ingrese la temperatura a convertir: "))

            if unidad_inicial == "C":
                if unidad_final == "K":
                    temperatura_convertida = temperatura + 273.15
                    print(f"Estimado {Nombre} la conversión de {temperatura:.2f} ºK es equivalente a {temperatura_convertida:.2f} ºK (Grados Kelvin)")
                elif unidad_final == "F":
                    temperatura_convertida = (temperatura * 1.8) + 32
                    print(f"Estimado {Nombre} la conversión de {temperatura:.2f} ºK es equivalente a {temperatura_convertida:.2f} ºF (Grados Fahrenheit)")
            elif unidad_inicial == "F":
                if unidad_final == "C":
                    temperatura_convertida = (temperatura - 32) / 1.8
                    print(f"Estimado {Nombre} la conversión de {temperatura:.2f} ºF es equivalente a {temperatura_convertida:.2f} ºC (Grados Celsius)")
                elif unidad_final == "K":
                    temperatura_convertida = ((temperatura - 32) * 5) / 9 + 273.15
                    print(f"Estimado {Nombre} la conversión de {temperatura:.2f} ºK es equivalente a {temperatura_convertida:.2f} ºK (Grados Kelvin)")
            elif unidad_inicial == "K":
                if unidad_final == "C":
                    temperatura_convertida = temperatura - 273.15
                    print(f"Estimado {Nombre} la conversión de {temperatura:.2f} ºK es equivalente a {temperatura_convertida:.2f} ºC (Grados Celsius)")
                elif unidad_final == "F":
                    temperatura_convertida = (temperatura - 273.15) * 1.8 + 32
                    print(f"Estimado {Nombre} la conversión de {temperatura:.2f} ºK es equivalente a {temperatura_convertida:.2f} ºF (Grados Fahrenheit)")

            if not pregunta_usuario():
                break

        except ValueError as e:     #--------(C)
            print(e)

if __name__ == "__main__":     
    conversor_temperatura()
```

```python
"""
EJERCICIO 3__________________________________________________________________
Contador de palabras de un texto.
Escribe una función que dado un texto, muestre las veces que aparece cada palabra.
"""
```

```python
# (A) Para empezar aplicamos con IMPORT el módulo 'string', proporciona varias funciones para la manipulación de cadenas
# (B) función limpiador_texto: el proceso de limpieza implica eliminar la puntuación y dividir el texto en una lista de palabras
    # usando el metodo (maketrans) y la constante incorporada (string.punctuation) para elimina cualquier símbolo de puntuación.
    # y reemplazar por una cadena vacía, con método (translate) se traduce y pasa a una nueva variable (texto_sin_puntuacion)
# (C) contador_palabras: establecemo una lista donde se almacenaran como resultado de un bucle (FOR) para cada palabra
    # con una condicional (IF) donde se itera y contabiliza cada palabra dentro del bucle.
# (D) impresion_conteo_palabras: esta es la etapa de impresión, de las palabras ingresados, contabilizadas con una diagramación
    # específica para tener los resultados mas legibles
# (E) main: este apartado genera un orden de ejecución, empezando con el (INPUT), 'texto_depurado' que ejcuta la funcion 
    # 'limpiador_texto', la variable 'lista_palabras' ejecuta la función contador_palabras y la función'impresion_conteo_palabras'.

import string     #--------(A)
print("-" * 42)
print(">>> BIENVENIDO AL CONTADOR DE PALABRAS <<<")
print("-" * 42)
nombre = input("Para proceder indíquenos su nombre: ")
Nombre = nombre.title()

def limpiador_texto(text):     #--------(B)
    quitar_puntuacion = str.maketrans('', '', string.punctuation)
    texto_sin_puntuacion = text.translate(quitar_puntuacion)
    t = texto_sin_puntuacion.split()
    return t

def contador_palabras(t):     #--------(C)
    lista_palabras = {}
    for palabra in t:
        if palabra not in lista_palabras:
            lista_palabras[palabra] = 0
        lista_palabras[palabra] += 1
    return lista_palabras

def impresion_conteo_palabras(lista_palabras):     #--------(D)
    print("-" * 24)
    print("Estimado/a " + (Nombre) +" aqui los resultados:")
    print("{:9} {:5}".format("Palabra", "Cantidad"))
    print("-" * 24)
    for palabra, cantidad in lista_palabras.items():
        print("{:9} {:5}".format(palabra, cantidad))

def main():                                      #--------(E)
    text = input('Introduce el texto que deseas procesar:').lower()
    texto_depurado = limpiador_texto(text)
    lista_palabras = contador_palabras(texto_depurado)
    impresion_conteo_palabras(lista_palabras)

main()
```

```python
"""
EJERCICIO 2__________________________________________________________________
Conversión de tipos de datos
El cliente recibe una lista de datos y necesita generar dos listas, la primera donde estarán 
todos los elementos que pudieron convertirse en flotantes y la otra donde están los elementos que 
no pudieron convertirse. Ejemplo de la lista que recibe el cliente: 
[ '1.3', 'one' , '1e10' , 'seven', '3-1/2', ('2',1,1.4,'not-a-number' ), [1,2,'3','3.4']]
"""
```

```python
# EXPLICACIÓN:
# (A)(INPUT) solicitamos nombre del usuaro con (INPUT) y establecemos una leve presentación de la aplicación
# agregamos la lista solicitada en el ejercicio
# (B)definimos 2 listas conforme el enunciado:
    # elementos_convertidos = []
    # elementos_no_convertidos = []
# (C)(FOR) Elaboramos un bucle para recorrer la lista inicial (cliente) Y definir la 
    # esclusión de errores tipos por medio de 'try - except', asi los elementos convertidos iran a 
    #'elementos_convertidos = []' y los elementos no convertidos se dirigen a otro bucle for, 
# (D)(TRY)nuevamente por medio de 2'try - except' para definir los elementos que se pueden convertir
    # se agregan a elementos_convertidos.
    # (EXCEPT) cuando los valores no son aptos conforme un TypeError y ValueError respectivamente a
    # los que no se pudieron convertir en float a la lista 'elementos_no_convertidos.
# (PRINT) mostramos el resultado en un orden convenido, desde nombre de usuario, lista de datos
    # convertirdos y no convertidos

print('-'*37)
print('>>>> CONVERSION DE TIPO DE DATOS <<<<')
print('      flotantes y no flotantes')
print('-'*37)
name = input("Cual es su nombre?: ")     #--------(A)
Name = name.title()
print('-'*37)

lista_inicial = [ '1.3', 'one' , '1e10' , 'seven', '3-1/2', ('2',1,1.4,'not-a-number' ), [1,2,'3','3.4']]
print(f'La lista inicial del cliente es: {lista_inicial}')

elementos_convertidos = []               #--------(B)
elementos_no_convertidos = []

for elemento in lista_inicial:           #--------(C)
    try:                                 #--------(D)
        elemento = float(elemento)
        elementos_convertidos.append(elemento)
    except TypeError: 
        try: 
            for e in elemento:
                e = float(e)
                elementos_convertidos.append(e)
        except ValueError: 
            elementos_no_convertidos.append(e)
    except ValueError:
        elementos_no_convertidos.append(elemento)
print(f'--------Estimado/a {Name} aqui presentamos los resultados--------- ')
print(f'Los datos convertidos en flotantes son: {elementos_convertidos}')
print(f'Los datos no convertidos en flotantes son: {elementos_no_convertidos}')
```

```python
"""
Ejercicio 1: Contador y ordenador de palabras de un texto
El cliente quedó contento con el contador de palabras, pero ahora quiere leer archivos TXT y que 
calcule la frecuencia de cada palabra ordenadas dentro de las entradas habituales del diccionario 
según la letra con la que comienzan, es decir, las claves deben ir de la A a la Z y dentro de la A 
debemos ir de la A a la Z.
"""
```

```python
#(A)(INPUT) recoge el nombre del usuario y la bienvenida
#(B)función contador: aqui se genera la entrada de palabras a procesar (t) se crea un 
    # diccionario vacío llamado diccionario_ordenado.
#(C)(FOR) bucle que itera sobre cada (palabra) en la lista de entrada t.
    #(fl = palabra[0]:) se extrae la primera letra (fl) de la (palabra) actual
    #(if fl not in...) condición verifica si la primera letra ( fl) existe como clave en
        # el archivo diccionario_ordenado, si no es así, entonces es la primera palabra que
        # comienza con esa letra.
    #(diccionario_ordenado[fl] = {}:) si la primera letra es nueva, se crea un subdiccionario vacío
        # para esa nueva letra, aqui se almacenará las frecuencias de las palabras con esa letra.
    #(if palabra not in diccionario_ordenado[fl]:) esta condición verifica si (palabra) actual
        # ya existe como clave dentro del subdiccionario de su primera letra ( fl).
        # si no es así, significa que esta es la primera vez que aparece esa palabra.
    #(diccionario_ordenado[fl][palabra] = 0:) si la palabra es nueva, esta línea inicializa la 
        # frecuencia de la palabra a 0 dentro del subdiccionario.
    #(diccionario_ordenado[fl][palabra] +=1:) si la palabra es nueva o no, se incrementa 
        #el recuento de frecuencia de la palabra (palabra) en 1.
#(D)(FOR)este bucle anidado itera sobre las claves (primeras letras) del archivo diccionario_ordenado
    #Para cada tecla ( fl), se ordena los pares de palabras-frecuencia dentro del subdiccionario
    # 'diccionario_ordenado[fl]' usando la sorted()función, los pares ordenados se vuelven a convertir
    # en un diccionario usando dict(), asi las frecuencias de palabras dentro de cada grupo de primeras
    # letras estén ordenadas alfabéticamente.
    # en el siguiente (sorted) se ordena el diccionario principal según sus claves (primeras letras) usando la sorted()función
    #los pares clave-valor ordenados se vuelven a convertir en un diccionario usando dict(), esto
    #define un orden alfabetico.
#(F)función para separar las palabras con SPLIT()
#(G)función para eliminar los signos de puntuacion u otros que pudiera contener el texto procesado
#(H)función main: este apartado genera un orden de ejecución con WHILE, usando TRY se ingresa el (INPUT), 
    # de la ruta del txt, usando (WITH open) se abre (.read) se lee el archivo para generar 'depurador'
    # 'separador' , 'contador' y la función de impresión personalizada incluyendo el nombre de usuario.
#(I)en este punto, definimos si la ruta ingresada no es correcta o no valida, en ese caso retornamos al inicio

print('-'*42)
print('>>>> CONTADOR Y ORDENADOR DE PALABRAS <<<<')
print('-'*42)
name = input("Estimado usuario, cual es su nombre?: ")
Name = name.title()
print('-'*42)
import string

def contador(t):          #-------------------------------------(B)
    diccionario_ordenado = {}
    for palabra in t:     #-------------------------------------(C)
        letra1 = palabra[0]
        if letra1 not in diccionario_ordenado:
            diccionario_ordenado[letra1] = {}
        if palabra not in diccionario_ordenado[letra1]:
            diccionario_ordenado[letra1][palabra] = 0
        diccionario_ordenado[letra1][palabra] +=1
    for key in diccionario_ordenado:     #-----------------------(D)
        diccionario_ordenado[key] = dict(sorted(diccionario_ordenado[key].items()))
    diccionario_ordenado = dict(sorted(diccionario_ordenado.items()))
    return diccionario_ordenado

def separador(t):         #---------------------------------------(F)             
    palabra = t.split()
    return palabra

def depurador(t):         #---------------------------------------(G)
    depurar1 = str.maketrans('','', string.punctuation)
    depurar2 = str.maketrans('','','¡')
    depurar3 = str.maketrans('','','¿')
    
    t1 = t.translate(depurar1)
    t2 = t1.translate(depurar2)
    t3 = t2.translate(depurar3)
    return t3

def main():      #-----------------------------------------------(H)
    while True:
        try:
            ruta = input("Ingrese la ruta del archivo 'txt': ")
            with open(ruta, 'r') as f:
                txt = f.read()

            t = txt.lower()
            t = depurador(t)
            t = separador(t)
            t = contador(t)

            diccionario_ordenado = sorted(t.items())
            print('-'*72)
            print(f'          Estimado/a {Name} aqui presentamos los resultados            ')
            print('-'*72)
            for key, value in diccionario_ordenado:
                print(f"{key:1} : {value}")

            break
        except FileNotFoundError:    #---------------------------(I)
            print("Error: El archivo no se ha encontrado. Intente de nuevo.")

if __name__ == "__main__":
    main()
```
